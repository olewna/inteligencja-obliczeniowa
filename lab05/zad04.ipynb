### Importing all required libraries and the csv file

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

irys = pd.read_csv("iris.csv")
```

### Random test and train values

```python
all_inputs = irys[['sepal.length','sepal.width','petal.length','petal.width']].values
all_classes = irys['species'].values

(X_train, X_test, y_train, y_test) = train_test_split(all_inputs, all_classes, train_size=0.7, random_state=278779)
```

### Classification k-NN, k=5 with percentage accuracy and confusion matrix

```python
knn1 = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn1.fit(X_train, y_train)
y_pred1 = knn1.predict(X_test)

result1 = round(knn1.score(X_test, y_test)*10000)/100
print(str(result1) + "%")

cm1 = confusion_matrix(y_test,y_pred1,labels=knn1.classes_)
disp1 = ConfusionMatrixDisplay(confusion_matrix=cm1,display_labels=knn1.classes_)
disp1.plot()
```

### Classification k-NN, k=3 with percentage accuracy and confusion matrix

```python
knn2 = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn2.fit(X_train, y_train)
y_pred2 = knn2.predict(X_test)

result2 = round(knn2.score(X_test, y_test)*10000)/100
print(str(result2) + "%")

cm2 = confusion_matrix(y_test,y_pred2,labels=knn2.classes_)
disp2 = ConfusionMatrixDisplay(confusion_matrix=cm2,display_labels=knn2.classes_)
disp2.plot()
```

### Classification k-NN, k=11 with percentage accuracy and confusion matrix

```python
knn3 = KNeighborsClassifier(n_neighbors=11, metric='euclidean')
knn3.fit(X_train, y_train)
y_pred3 = knn3.predict(X_test)

result3 = round(knn3.score(X_test, y_test)*10000)/100
print(str(result3) + "%")

cm3 = confusion_matrix(y_test,y_pred3,labels=knn3.classes_)
disp3 = ConfusionMatrixDisplay(confusion_matrix=cm3,display_labels=knn3.classes_)
disp3.plot()
```

### Naive Bayes Classification with percentage accuracy and confusion matrix

```python
bayes = GaussianNB()
bayes.fit(X_train, y_train)
y_pred4 = bayes.predict(X_test)

result4 = round(bayes.score(X_test, y_test)*10000)/100
print(str(result4) + "%")

cm4 = confusion_matrix(y_test,y_pred4,labels=bayes.classes_)
disp4 = ConfusionMatrixDisplay(confusion_matrix=cm4,display_labels=bayes.classes_)
disp4.plot()
```

### Taking every classification from all of those 5 (with the one in previous exercise) into consideration, it looks like the k-NN with k=11 and Naive Bayes algorithms are the best with 95.56% of accuracy.
