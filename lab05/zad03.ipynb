```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import metrics
import matplotlib.pyplot as plt

irys = pd.read_csv("iris.csv")

all_inputs = irys[['sepal.length','sepal.width','petal.length','petal.width']].values
all_classes = irys['species'].values

(train_inputs, test_inputs, train_classes, test_classes) = train_test_split(all_inputs, all_classes, train_size=0.7, random_state=278779)
```

```python
dtc = DecisionTreeClassifier()

dtc.fit(train_inputs, train_classes)

result = round(dtc.score(test_inputs, test_classes)*10000)/100
print(str(result) + "%")
```

```python
tree.plot_tree(dtc)
```

```python
dtc.fit(train_inputs, train_classes)
y_pred = dtc.predict(test_inputs)

cm = metrics.confusion_matrix(test_classes,y_pred,labels=dtc.classes_)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=dtc.classes_)
disp.plot()
```
