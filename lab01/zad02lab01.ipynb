```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#a
# reading to python file from csv file
miasta = pd.read_csv("miasta.csv")

#b
# creating DataFrame and adding it to new one by merging both
miasta2010 = pd.DataFrame([[2010,460,555,405]], columns=["Rok","Gdansk","Poznan","Szczecin"])
# print(miasta2010)
miasta = pd.concat([miasta,miasta2010], ignore_index=True)

#c
# created plot with mathplotlib using object values and added labels 
plt.plot(miasta["Rok"],miasta["Gdansk"], '-ro', label="Gdansk")
plt.ylabel('ludnosc (w tys)')
plt.xlabel('lata')
# plt.show()

#d
# added rest of cities with legend
plt.plot(miasta["Rok"],miasta["Poznan"],'-bo', label="Poznan")
plt.plot(miasta["Rok"],miasta["Szczecin"],'-go', label="Szczecin")
plt.legend()
plt.show()

#e
# created mean and std before standarization and after this we can see that mean is 0 and std is 1
mean = np.mean(miasta.values)
std = np.std(miasta.values)
afterSt = []

for i in miasta.values:
    x = (i-mean)/std
    afterSt.append(x)

# print(afterSt)
meanAfter = np.mean(afterSt)
stdAfter = np.std(afterSt)
print(round(meanAfter,2), stdAfter)

#f
# before normalizing we have lowests and biggest value and after the lowest is 0 and biggest is 1
minx = min(miasta.values.flatten())
maxx = max(miasta.values.flatten())
afterNor = []

for i in miasta.values:
    x = (i-minx)/(maxx-minx)
    afterNor.append(x)

min_after = np.min(afterNor)
max_after = np.max(afterNor)
print(min_after,max_after)
```
