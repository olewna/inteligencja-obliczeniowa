```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
diabetes = pd.read_csv("diabetes.csv")

all_inputs = diabetes[['pregnant-times','glucose-concentr','blood-pressure','skin-thickness','insulin','mass-index','pedigree-func','age']].values
all_classes = diabetes['class'].values

(train_inputs, test_inputs, train_labels, test_labels) = train_test_split(all_inputs, all_classes, train_size=0.7, random_state=278779)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(6,3),max_iter=500,random_state=278779, activation="relu")

mlp.fit(train_inputs, train_labels)

predictions_test = mlp.predict(test_inputs)
print(accuracy_score(predictions_test, test_labels))
```

```python
cm = confusion_matrix(test_labels, predictions_test)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=mlp.classes_)
disp.plot()
```
